**3. ให้นำข้อมูลจาก data_clean_bfclean.csv คอลัมน์ SQ_M มาจัดการ**

โดยที่
- ข้อมูลที่มีค่าน้อยกว่า 100 ถือว่าเป็นข้อมูลที่ผิดพลาด ให้เปลี่ยนเป็น Null
- ข้อมูลที่ไม่ใช่ตัวเลขถือเป็นข้อมูลที่ผิดพลาด ให้เปลี่ยนเป็น Null
- ทำคอลัมน์ให้เป็น float64

```python
dc = pd.read_csv('data_clean_bfclean.csv')
dc
```

```python
dc.info()
```

```python
def fn(x):
    try:
        val = float(x)
        if val <100:
            return np.nan
        else: return val
    except:
        return np.nan
dc['SQ_M'] = dc['SQ_M'].apply(fn)
```

```python
def fn(x):
    try:
        val = float(x)
        if val <100:
            return np.nan
        else: 
            return val
    except:
        return np.nan
        
dc['SQ_M'] = dc['SQ_M'].apply(fn).astype("Float64", errors="ignore")
dc
```
