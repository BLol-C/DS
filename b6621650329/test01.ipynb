```python
def printName(first,last,initials):
    if initials:
        print(first[0]+'.'+last[0]+'.')
    else:
        print(first,last)
```

```python
printName('Patchara','Prasantaveepol',True)
```

```python
printName('Patchara','Prasantaveepol',False)
```

```python
printName('Prasantaveepol','Patchara',False)
```

```python
printName('Prasantaveepol','Patchara')
```

```python
printName(last='Prasantaveepol',first='Patchara',False)
```

```python
printName('Patchara',initials=False,last='Prasantaveepol')
```

```python
printName(last='Prasantaveepol',first='Patchara',initials=False)
```

```python
def printName(first,last,initials=True):
    if initials:
        print(first[0]+'.'+last[0]+'.')
    else:
        print(first,last)
```
