```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dic = {'day': 'Sun Mon Tue Wed Thu Fri Sat'.split(),
        'trmp':[24,25,29,35,23,20,23]}
df = pd.DataFrame(dic)
df
```

```python
plt.figure(figsize = [4,3])
plt.plot(df['day'], df['trmp'])
plt.show()
```

```python
plt.figure(figsize = [4,3])
x = df['day']
y = df['trmp']
plt.plot(x, y, ls = '-.',
         lw = 1,
         c = (1,0.2,.5),
         marker = 's',
         ms = 5,
         mec = (1,0.2,.5),
         mfc = 'red')
plt.show()
```

```python
plt.figure(figsize = [4,3])
x = df['day']
y = df['trmp']
plt.plot(x,y,'*--r')
plt.show()
```

```python
dic = {'day': 'Sun Mon Tue Wed Thu Fri Sat'.split(),
       'temp': [24,25,29,35,23,20,23],
       'temp2': [10,14,9,15,20,15,50]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp']
z = df['temp2']
df
```

```python
plt.figure(figsize = [4,3])
plt.plot(x,y,'s--r', label = 'T1')
plt.plot(x,z,'s-b', label = 'T2')
plt.legend(loc = 'upper left')

plt.title('Test02')
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0,0].plot(x,y)
axes[0,0].set_title('Test01')

axes[0,1].plot(x,z)
axes[0,1].set_title('Test02')

fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows = 3, figsize = [5,8])
axes[0].plot(x,y)
axes[0].set_title('Test01')

axes[2].plot(x,z)
axes[2].set_title('Test02')

fig.tight_layout()
plt.show()
```
