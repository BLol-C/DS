# Data Visualization with Seaborn - Exercises 

## The Data

ใน exercise นี้ เราจะใช้ titanic data set ซึ่งเป็น Data ที่มีชื่อเสียงเป็นที่นิยม ซึ่งมีให้อยู่ใน library ของ seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

# Exercises

** ให้นำ data ของ titanic ที่เป็น dataframe ไปใช้ทำการ plot ให้ตรงกับภาพดังต่อไปนี้**

** *Note! เพื่อไม่ให้ภาพที่ถูกต้องหาย อย่าลืม insert cell ขึ้นมาข้างบนใหม่มาขั้นก่อนทำการ run code จะได้ไม่ทำให้เกิดกันทับซ้อน (overwrite)* **


**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.pairplot(data = titanic)
plt.show()
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
g= sns.jointplot(data = titanic, x = 'fare', y = 'age', hue = 'sex')

g.fig.set_size_inches(6,6)
plt.show()
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
plt.figure()

sns.histplot(data = titanic, x = 'fare', bins = 20, kde = True, color = 'r')

plt.show()
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
plt.figure()

sns.swarmplot(data = titanic, x = 'class', y = 'age', size = 4)

plt.show()
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
plt.figure()

sns.boxplot(data = titanic, x = 'sex', y = 'age', hue = 'class')

plt.show()
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
plt.figure(figsize=[5,4])

sns.countplot(data = titanic, x = 'sex', hue = 'class', palette = 'Set1')

plt.show()
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.heatmap(titanic.select_dtypes(include = ['int64', 'float64']).corr(), cmap = 'coolwarm', lw = 2)
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
g = sns.FacetGrid(data = titanic, col = 'class', row = 'sex')
g.map(sns.histplot, 'age', bins = 10)

plt.show()
```
