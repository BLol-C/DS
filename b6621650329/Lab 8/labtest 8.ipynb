```python
plt.figure(figsize=[4,3])

sns.barplot(data = df, x = 'day', y = 'tip', hue = 'day', legend = True)

plt.title('Test')
plt.show()
```

```python
joint_kws แก้ตรงกลาง
marginl_kws แก้รอบข้าง
```

```python
g = sns.jointplot(data = df, x = 'total_bill', y = 'tip',
                  kind = 'scatter',
                  joint_kws = {'color':'r','marker':'s'},
                  marginal_kws = {'kde':'True','bins':20})

g.fig.set_size_inches(3,3)
plt.show()
```

```python
g = sns.JointGrid(data = df, x = 'total_bill', y = 'tip')
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.boxplot)


g.fig.set_size_inches(3,3)
plt.show()
```

```python
diag_kws กำหนดมุมแนวทแยง
```

```python
g = sns.pairplot(data = df, hue = 'sex')
g.fig.set_size_inches(5,5)
```

```python
g = sns.PairGrid(data = df)
g.map_diag(sns.histplot)
g.map_upper(sns.lineplot, marker = 's')
g.map_lower(sns.scatterplot, color = 'r')

g.fig.set_size_inches(5,5)
plt.show()
```

```python
g = sns.FacetGrid(data = df, col = 'day', row = 'smoker', hue = 'sex')
g.map(sns.scatterplot, 'total_bill', 'tip')

g.add_legend()
plt.show()
```
