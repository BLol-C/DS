```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from mlxtend.frequent_patterns import apriori , association_rules
```

```python
raw_data = pd.read_excel('./data/MarketBasketOptimization.xlsx',header = None)
#openpyxl
```

```python
raw_data
```

```python
dataset = []
for i in range(0 , len(raw_data)):
    tmp = []
    for j in range(0 , len(raw_data.loc[i])) :
        if raw_data.loc[i,j] is not np.nan :
            tmp.append(raw_data.loc[i,j])
    dataset.append(tmp)
```

```python
dataset[3]
```

```python
from mlxtend.preprocessing import TransactionEncoder
TE = TransactionEncoder()
tmp_dataset = TE.fit_transform(dataset)
```

```python
tmp_dataset
```

```python
df = pd.DataFrame(tmp_dataset , columns = TE.columns_)
df.head(2)
```

```python
itemset = apriori(df.drop('mineral water',axis=1), min_support=0.01, use_colnames=True )
itemset
```

```python
rule = association_rules( itemset , metric='confidence', min_threshold=0.30 )
rule
```
